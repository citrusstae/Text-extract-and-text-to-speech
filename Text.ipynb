```python
text= """The diagram illustrates an underwater explorer which is remotely controlled from a ship on the surface. The explorer has many device attached with it for exploration tasks.

Firstly, the underwater explorer is 2.7m in length and 1.8m in width as well as in height. The explorer can reach the top speed of 3 knots with weight capacity of 3,200kg. The explorer has maximum depth at 4,500m from ship.

The explorer is equipped  with multiple devices which helps to carry out the exploration of ocean bed. The device contains two cameras at centre. Both cameras are installed with specific purpose. HD video camera captures the video footage for the monitor at the ship whereas Digital still camera is used for capturing still images. Although the explorer is guided by monitor on the ship a position beacon is attached with the device to know the location of the explorer. Next to video cameras two led light on either side are attach for clear visibility. With LED lights a pair of laser lights are also present on the explorer. Interestingly, to move underwater a pair of thruster is used.

Lastly, the explorer is guided by a person from the ship and explorer is currently 0.74m away from the ocean floor.


better response:

The diagram illustrates a remotely controlled underwater explorer used for ocean bed exploration, which is operated from a ship on the surface. Several devices are attached to the explorer to assist in its tasks.

In terms of its dimensions, the explorer measures 2.7 meters in length and 1.8 meters in both width and height. It has a maximum speed of 3 knots and can carry a load of up to 3,200 kilograms. The explorer is capable of descending to a depth of 4,500 meters.

For exploration, the explorer is equipped with two cameras located at its center. One is an HD video camera, which transmits live footage to the surface shipâ€™s monitor, while the other is a digital still camera used for capturing images. Additionally, a position beacon is attached to help track the explorer's location. On either side of the cameras, there are two LED lights to enhance visibility, alongside a pair of laser lights.

To navigate underwater, the explorer uses a pair of thrusters. It is currently positioned 0.74 meters above the ocean floor and is remotely operated by personnel on the surface."""
```

```python
len(text)
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation ## used for text cleaning
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc=nlp(text)
```

```python
tokens = []

for token in doc:
    # Check if the token is not a stop word, not punctuation, and not a newline character
    if not token.is_stop and not token.is_punct and token.text != '\n':
        tokens.append(token.text.lower())

# Now `tokens` contains the filtered and lowercased tokens
```

```python
tokens
```

```python
# pos tagging
token1=[]
stopwords= list(STOP_WORDS)
allowed_pos = ['ADJ','PROPN','VERB','NOUN']
for token in doc:
    if token.text in stopwords or token.text in punctuation:
        continue
    if token.pos_ in allowed_pos:
            token1.append(token.text)
```

```python
token1
```

```python
# calculate frequency of each words
from collections import Counter # this function counts the frequency of words
```

```python
word_fre=Counter(tokens)
```

```python
word_fre
```

```python
max_freq= max(word_fre.values())
```

```python
max_freq
```

```python
#normalization
for word in word_fre.keys():
    word_fre[word] = word_fre[word]/max_freq
```

```python
word_fre
```

```python
#sentence tokenisation
sent_token=[sent.text for sent in doc.sents]
```

```python
sent_token
```

```python
sent_score= {}
for sent in sent_token:
    for word in sent.split():
        if word.lower() in word_fre.keys():
            if sent not in sent_score.keys():
                sent_score[sent]=word_fre[word]
            else:
                sent_score[sent]+=word_fre[word]
        print(word)
```

```python
sent_score
```

```python
import pandas as pd
```

```python
pd.DataFrame(list(sent_score.items()),columns=['Sentence','Score'])
```

```python
from heapq import nlargest
```

```python
num_sentences = 5
n = nlargest(num_sentences,sent_score,key=sent_score.get)
```

```python
from transformers import pipeline
```

```python
summarizer=pipeline("summarization",model='t5-base',tokenizer='t5-base',framework='pt')
```

```python
summary=summarizer(text,max_length=500,min_length=10,do_sample=False)
```
